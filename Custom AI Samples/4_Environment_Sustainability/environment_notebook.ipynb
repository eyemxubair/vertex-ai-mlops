# Environment & Sustainability: Pollution Analysis and Alerts

```python
import pandas as pd
df = pd.read_csv('pollution_readings.csv', parse_dates=['timestamp'])
daily = df.groupby([df['timestamp'].dt.date,'sensor_id'])[['pm25','no2','o3']].mean().reset_index()
daily.head()
```

## IsolationForest for Anomaly Detection

```python
from sklearn.ensemble import IsolationForest
sample = daily[['pm25','no2','o3']].values
iso = IsolationForest(random_state=42, contamination=0.02).fit(sample)
daily['anomaly'] = iso.predict(sample)
alerts = daily[daily['anomaly']==-1].copy()
alerts[['sensor_id']].value_counts().head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,3))
plt.plot(daily.index, daily['pm25'])
plt.scatter(alerts.index, alerts['pm25'], s=10)
plt.title('PM2.5 with Alerts'); plt.show()
```
