```
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
DATA_DIR = Path(r'/mnt/data/Customer-Service/data')
complaints = pd.read_csv(DATA_DIR/'complaints_train.csv')
complaints.head()
```

```
complaints['text_len']=complaints['text'].str.len()
complaints.groupby('label').size().plot(kind='bar', title='Class Counts')
plt.show()
```

```
from sklearn.model_selection import train_test_split
train_df, val_df = train_test_split(complaints, test_size=0.2, stratify=complaints['label'], random_state=42)
print(train_df.shape, val_df.shape)
train_df.to_csv('train_split.csv',index=False)
val_df.to_csv('val_split.csv',index=False)
print('Saved train_split.csv and val_split.csv')
```
