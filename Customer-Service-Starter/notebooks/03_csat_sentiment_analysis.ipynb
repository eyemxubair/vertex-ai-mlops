# 03 â€“ CSAT & Sentiment Analysis

```
import pandas as pd
import matplotlib.pyplot as plt
complaints = pd.read_csv('val_split.csv')
csat = pd.read_csv(r'/mnt/data/Customer-Service/data/csat_surveys.csv')
merged = csat.merge(complaints[['complaint_id','label','sentiment']], left_on='interaction_id', right_on='complaint_id', how='left')
merged.groupby('label')['rating_1_5'].mean().plot(kind='bar', title='Avg CSAT by Issue label')
plt.show()
merged.head()
```

```
# VADER baseline sentiment for CSAT comments
# !pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
csat['vader_score'] = csat['comment'].apply(lambda t: analyzer.polarity_scores(str(t))['compound'])
csat[['comment','vader_score']].head()
```

```
csat['is_promoter'] = csat['nps']>=9
by_channel = csat.groupby('channel')['is_promoter'].mean().sort_values(ascending=False)
by_channel.plot(kind='bar', title='Promoter Rate by Channel')
plt.show()
```
